```python
import pandas as pd
import os
```

```python
## Read in Accident CSV file
file = os.path.join("..","Cleaned", "2014_2018_merged.csv")
```

```python
data = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
data.columns
```

```python
df = data.drop(columns=['Unnamed: 0', 'ST_CASE', 'PERSONS', 'COUNTY', 'CITY', 'DAY','YEAR','MINUTE', 'ROUTE', 'TWAY_ID', 
    'TWAY_ID2','LATITUDE', 'LONGITUD','WEATHER1', 'WEATHER2', 'CF1', 'CF2', 'CF3', 'FATALS','PER_TYP', 'TRAV_SP','Year', 'ROAD_FNC'])
```

```python
df.head()
```

```python
hours = []

for i in df['HOUR']:
    if i > 0 and i < 6:
        hours.append('Early Morning')
    elif i < 12:
        hours.append('Mid Morning')
    elif i < 18:
        hours.append('Mid Afternoon')
    else:
        hours.append('Party Time')

df['Time Group']= hours
df = df.drop(columns=['HOUR'])
df.head()
```

```python
inj = []

for i in df['INJ_SEV']:
    if i == 0:
        inj.append('0')
    elif i > 0 and i < 4:
        inj.append('1')
    elif i == 4:
        inj.append('2')
    else:
        inj.append('0')


df = df.drop(columns=['INJ_SEV'])
df['INJ_SEV']= inj
df.head()
```

```python
drugs = []

for i in df ['DRUGS']:
    if i == 0 or i >1:
        drugs.append(0)
    else:
        drugs.append(1)


df = df.drop(columns=['DRUGS'])
df['DRUG_USE'] = drugs
df.head()
```

```python
drinks = []

for i in df['DRINKING']:
    if i == 0 or i > 1:
        drinks.append(0)
    else:
        drinks.append(1)

df = df.drop(columns=['DRINKING'])
df['DRINKING'] = drinks
df.head()
```

```python
weather = []

for i in df['WEATHER']:
    if i == 8 or i > 10:
        weather.append(8)
    else:
        weather.append(i)
        
df = df.drop(columns=['WEATHER'])
df['WEATHER'] = weather
df.head()
```

```python
helmet = []

for i in df['REST_USE']:
    if i == 19 or i == 5 or i == 6 or i == 16 or i ==17:
        helmet.append(1)
    else:
        helmet.append(0)
        
df = df.drop(columns=['REST_USE'])
df['HELMET'] = helmet
df.head()
```

```python
intersections = []

for i in df['TYP_INT']:
    if  i ==1 or i == 8 or i == 9 or i >11:
        intersections.append(0)
    else:
        intersections.append(i)
        
df = df.drop(columns=['TYP_INT'])
df['Intersection'] = intersections
df.head()
```

```python
list(df.columns)
```

```python
cleaned = pd.get_dummies(df,columns=['MONTH', 'DAY_WEEK','SEX','DRUG_USE','DRINKING','WEATHER','Intersection'])
# 'INJ_SEV',
```

```python
cleaned.columns
```

```python
new_df = cleaned.rename(columns={'MONTH_1':'January','MONTH_2':'February','MONTH_3':'March','MONTH_4':'April','MONTH_5':'May','MONTH_6':'June',
                       'MONTH_7':'July','MONTH_8':'August','MONTH_9':'September','MONTH_10':'October','MONTH_11':'November',
                        'MONTH_12':'December','DAY_WEEK_1':'Sunday','DAY_WEEK_2':'Monday','DAY_WEEK_3':'Tuesday','DAY_WEEK_4':'Wednesday',
                       'DAY_WEEK_5':'Thursday','DAY_WEEK_6':'Friday','DAY_WEEK_7':'Saturday','SEX_1':'Male','SEX_2':'Female','SEX_9':'Gender Neutral',
                        'INJ_SEV_0':'No Injury','INJ_SEV_1':'Injured','INJ_SEV_2':'Fatality','DRUG_USE_0':'No Drugs',
                        'DRUG_USE_1':'Drug Use Reported','DRINKING_0':'No Drinking','DRINKING_1':'Drinking Reported',
                       'WEATHER_1':'Clear Weather','WEATHER_2':'Rainy Weather','WEATHER_5':'Foggy Weather', 
                        'WEATHER_6':'Severly Windy Weather','WEATHER_7':'Blowing Sand Weather','WEATHER_8':'Other Weather',
                        'WEATHER_10':'Cloudy Weather', 'AGE': 'Age', 'HELMET': 'Helmet Use', 'PREV_ACC': 'Previous Accident', 
                        'PREV_DWI': 'Previous DWI', 'PREV_SPD': 'Previous Speeding', 'Intersection_0': 'No Intersection',
                        'Intersection_2':'Four-way Intersection', 'Intersection_3':'T-Intersection','Intersection_4':'Y-Intersection', 'Intersection_5': 'Traffic Circle',
                        'Intersection_7':'Five-point Intersection','Intersection_10':'L-Intersection'})
```

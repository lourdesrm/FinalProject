## Merge files on ST_CASE each year

```python
import os
import pandas as pd
from functools import reduce
```

```python
# Import previously cleaned files
# Accidents
accident_2018 = os.path.join("..", "Cleaned","Accident", "2018_Accident.csv")
accident_2017 = os.path.join("..", "Cleaned","Accident", "2017_Accident.csv")
accident_2016 = os.path.join("..", "Cleaned","Accident", "2016_Accident.csv")
accident_2015 = os.path.join("..", "Cleaned","Accident", "2015_Accident.csv")
accident_2014 = os.path.join("..", "Cleaned","Accident", "2014_Accident.csv")

# Person
person_2018 = os.path.join("..", "Cleaned","Person", "person_2018.csv")
person_2017 = os.path.join("..", "Cleaned","Person", "person_2017.csv")
person_2016 = os.path.join("..", "Cleaned","Person", "person_2016.csv")
person_2015 = os.path.join("..", "Cleaned","Person", "person_2015.csv")
person_2014 = os.path.join("..", "Cleaned","Person", "person_2014.csv")

# Vehicle
vehicle_2018 = os.path.join("..", "Cleaned","Vehicle", "2018_Vehicle.csv")
vehicle_2017 = os.path.join("..", "Cleaned","Vehicle", "2017_Vehicle.csv")
vehicle_2016 = os.path.join("..", "Cleaned","Vehicle", "2016_Vehicle.csv")
vehicle_2015 = os.path.join("..", "Cleaned","Vehicle", "2015_Vehicle.csv")
vehicle_2014 = os.path.join("..", "Cleaned","Vehicle", "2014_Vehicle.csv")
```

```python
# Read imported files
# Accidents
accident_2018_df=pd.read_csv(accident_2018)
accident_2017_df=pd.read_csv(accident_2017)
accident_2016_df=pd.read_csv(accident_2016)
accident_2015_df=pd.read_csv(accident_2015)
accident_2014_df=pd.read_csv(accident_2014)

# Person
person_2018_df=pd.read_csv(person_2018)
person_2017_df=pd.read_csv(person_2017)
person_2016_df=pd.read_csv(person_2016)
person_2015_df=pd.read_csv(person_2015)
person_2014_df=pd.read_csv(person_2014)

# Vehicle
vehicle_2018_df=pd.read_csv(vehicle_2018)
vehicle_2017_df=pd.read_csv(vehicle_2017)
vehicle_2016_df=pd.read_csv(vehicle_2016)
vehicle_2015_df=pd.read_csv(vehicle_2015)
vehicle_2014_df=pd.read_csv(vehicle_2014)
```

## MERGE DFs

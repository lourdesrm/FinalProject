```python
import pandas as pd
import os
```

```python
## Read in Accident CSV file
file = os.path.join("..","Raw_files", "FARS2003NationalCSV","ACCIDENT.csv")
```

```python
# create dataFrame and load in data from CSV file
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
## Create a new dataFrame for just Texas cases since data is over 100k rows
tx_df = df.loc[df['STATE'] == 48]
```

```python
## Dropping Unneccesary Columns
clean = tx_df.drop(columns=['STATE','VE_FORMS','PEDS','NHS','MILEPT','SP_JUR','HARM_EV','MAN_COLL','REL_ROAD','LGT_COND','SCH_BUS','RAIL','NOT_HOUR','NOT_MIN','ARR_HOUR','ARR_MIN','HOSP_HR','HOSP_MN','DRUNK_DR',])
#,'RUR_URB','FUNC_SYS','RD_OWNER','PVH_INVL','PERNOTMVIT','PERMVIT','RELJCT1','RELJCT2','WRK_ZONE','VE_TOTAL',
```

```python
## Check new DF to ensure file read accurately
clean
```

```python
## Create new CSV with cleaned TX data
clean.to_csv('../Cleaned/Accident/2003_Accident.csv', index=False)
```

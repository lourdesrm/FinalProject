```python
import pandas as pd
import os
```

```python
## Read in Accident CSV file
file = os.path.join("..","Cleaned", "testing_merged_sub.csv")
```

```python
data = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
list(data.columns)
```

```python
cleaned = pd.get_dummies(data,columns=['MONTH_x', 'DAY_WEEK', 'TYP_INT', 'WEATHER','SEX'])
```

```python
hours = []

for i in data['HOUR_x']:
    if i > 0 and i < 6:
        hours.append('Early Morning')
    elif i < 12:
        hours.append('Mid Morning')
    elif i < 18:
        hours.append('Mid Afternoon')
    else:
        hours.append('Party Time')
    
```

```python
data['Time Group']= hours
```
